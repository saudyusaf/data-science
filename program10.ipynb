```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
values = np.random.randn(100)
s5 = pd.Series(values) # generate a pandas series
print("series\n",s5)
s5.plot(kind='hist', title='Normally distributed random values')
plt.show()
```
